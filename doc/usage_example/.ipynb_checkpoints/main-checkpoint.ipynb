# Created base table
select application_id as appl_id,

year_quarter,

case

when segment = 'АК' then 'auto'

when segment = 'ИК' then 'mort'

when segment = 'КК' then 'cc'

when segment = 'ПК' then iif(exist = 0, 'pil_new', 'pil_exist')

end as segment,

target

into Modeling.dbo.bki_base

from  Modeling.dbo.application 

where application_id % 10 <> 0

```python
data = fe.read_data(ch=ch,
                    base_schema_table='Modeling.dbo.bki_base',
                    base_table_id = 'appl_id',
                    process='appl',
                    reducing_memory_func=partial(MemoryReducer(float_min_type=32, int_min_type=8).reduce, verbose=1))
```

```python
not_features = ['appl_id', 'year_quarter', 'segment', 'target']
features = [f for f in data.columns if f not in not_features]
target = 'target'
grp = 'segment'
groupby = ['year_quarter', 'segment']
data
```

```python
feats_summary = Summary(fe.feature_description).features_summary(df=data,
                                                                 features_list=features,
                                                                 target_name=target,
                                                                 groupby_name=grp)
feats_summary.to_excel('report/feats_summary.xlsx')
feats_summary
```

# Сводка по переменным для отчета помещена в report/feats_summary.xlsx
CFP0[1-3]_CFT05 - все None

```python
not_monotonic = ['type_pil', 'type_mort', 'type_auto']
monotonic = {}
for f in features:
    if f in not_monotonic:
        monotonic[f] = 0
    elif feats_summary.loc[f]['score'] > 0:
        monotonic[f] = 1
    else:
        monotonic[f] = -1
```

```python
for f in features:
    if feats_summary.loc[f]['na_count'] == 0:
        continue
    best_sc = feats_summary.loc[f]['score_avg']
    if feats_summary.loc[f]['score_avg_na_min'] == best_sc:
        data[f] = data[f].fillna(feats_summary.loc[f]['min'] - 1)
    elif feats_summary.loc[f]['score_avg_na_max'] == best_sc:
        data[f] = data[f].fillna(feats_summary.loc[f]['max'] + 1)
    elif feats_summary.loc[f]['score_avg_na_mean'] == best_sc:
        data[f] = data[f].fillna(feats_summary.loc[f]['mean'])
    elif feats_summary.loc[f]['score_avg_na_median'] == best_sc:
        data[f] = data[f].fillna(feats_summary.loc[f]['median'])
    elif feats_summary.loc[f]['score_avg_na_mode'] == best_sc:
        data[f] = data[f].fillna(feats_summary.loc[f]['mode'])
```

```python
data.to_pickle('data.pkl', compression='gzip')
```

```python
selected1 = [f for f in features if abs(feats_summary.loc[f]['nunique']) > 1]
selected2 = [f for f in selected1 if abs(feats_summary.loc[f]['mode_freq']) < 0.95]
selected3 = [f for f in selected2 if abs(feats_summary.loc[f]['score']) > 0.01 or f in not_monotonic]
selected4 = [f for f in selected3 if abs(feats_summary.loc[f]['score_avg']) > 0.01 or f in not_monotonic]
print(len(features), len(selected1), len(selected2), len(selected3), len(selected4))
```

```python
params = {
    'max_iter': 1000,
    'early_stopping_rounds': 100,
    'max_depth': 5,
    'metric': 'auc',
    'boost': 'gbdt',
    'objective': 'binary'
}
models = []
skf = StratifiedStratifiedKFold(5, True, 16777216)
for itr, iva in skf.split(data[selected], data[target], data[groupby]):
    xtr, ytr = data.iloc[itr][selected], data.iloc[itr][target]
    xva, yva = data.iloc[iva][selected], data.iloc[iva][target]
    trn, val = lgb.Dataset(xtr, ytr), lgb.Dataset(xva, yva)
    model = lgb.train(params, trn, 1000, [trn, val], verbose_eval=50)
    models += [model]
```

```python
supermodel = join_lgb_boosters(models)
```

```python
pred = np.mean([model.predict(data[selected], raw_score=True) for model in models], axis=0)
superpred = supermodel.predict(data[selected], raw_score=True)
print(pred)
print(superpred)
```
